```python
# asignando la url a una variable
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

<p>Los datos que necesitas extraer son el nombre del lenguaje de programaci칩n y el salario medio anual. Es una buena idea abrir la URL en tu navegador web y estudiar el contenido de la p치gina web antes de empezar a extraer datos.</p>

```python
# importando laslibrerias necesarias
from bs4 import BeautifulSoup # Este modulo ayuda con el web scrapping.
import requests  # Este modulo ayuda para descargar la webpage
import pandas as pd
```

```python
# Descargando la webpage de la url
data  = requests.get(url).text
```

```python
# creando un objeto soup
soup = BeautifulSoup(data,"html.parser")
```

```python
# Scrape el Lenguaje de programacion y  promedio anual de salario
#Encontrando la tabla html en la web page
table = soup.find('table') # En html la tabla es representado por la etiqueta<table>
```

```python
# Obteniendo todas las filas de la tabla
for row in table.find_all('tr'): # En html table fila es representado por la etiqueta <tr>
    # Obteniendo todas las columnas en cada fila.
    cols = row.find_all('td') # in html a column is represented by the tag <td>
    language = cols[1].getText() # store the value in column 3 as color_name
    creator = cols[2].getText() # store the value in column 4 as color_code
    annual_salary = cols[3].getText() # store the value in column 4 as color_code
    difficult = cols[4].getText() # store the value in column 4 as color_code
    print("{}--->{}---->{}--->{}".format(language, creator, annual_salary, difficult))
```

```python
# convirtiendo la taba html a un dataframe de pandas
df = pd.read_html(str(table))[0]
```
